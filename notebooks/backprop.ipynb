### Пример: логистическая регрессия на Pytorch

Обучим простейшую модель — логистическую регрессию — на случайных данных, чтобы увидеть, как работает градиентный спуск в Pytorch.

**Модель** — линейная комбинация признаков, пропущенная через сигмоиду:

$\hat{y} = \sigma(w^T x + b), \quad \sigma(t) = \frac{1}{1 + \exp(-t)}$

**Функция ошибки** — бинарная кросс-энтропия:

$\text{CE}(y, \hat{y}) = -y \cdot \log \hat{y} - (1 - y) \log (1 - \hat{y})$

<img src="https://pytorch.org/tutorials/_images/comp-graph.png" style="background:white" width="600"/>

```python
import torch
```

```python
torch.manual_seed(42)
n_samples = 16
n_features = 5
x = torch.randn(n_samples, n_features)  # входной тензор
y = torch.randint(2, size=(n_samples, 1)).float()  # выходной тензор
w = torch.randn(
    n_features, 1, requires_grad=True
)  # параметр, хотим обновлять градиентным спуском
b = torch.randn(1, requires_grad=True)  # параметр, хотим обновлять градиентным спуском
```

Прямой проход: вычисляем предсказания и значение ошибки

```python
z = torch.matmul(x, w) + b
y_hat = torch.sigmoid(z)
loss = (-y * y_hat.log() - (1 - y) * (1 - y_hat).log()).mean()
loss
```

Pytorch запомнил все операции и построил вычислительный граф:

```python
print(f"Gradient function for z = {z.grad_fn}")
print(f"Gradient function for loss = {loss.grad_fn}")
```

В Pytorch уже есть готовая реализация этой функции ошибки — `binary_cross_entropy_with_logits` (принимает логиты `z` напрямую, без применения сигмоиды):

```python
loss_builtin = torch.nn.functional.binary_cross_entropy_with_logits(z, y)
print(f"Вручную:  {loss:.4f}")
print(f"Встроенная: {loss_builtin:.4f}")
```

Напишем цикл для поиска параметров, минимизирующих функцию ошибки

```python
n_iter = 1000
step = 0.1
for i in range(n_iter):
    z = x @ w + b
    loss = torch.nn.functional.binary_cross_entropy_with_logits(z, y)
    # посчитаем производные
    loss.backward()

    # обновим значения параметров внутри менеджера контекста `no_grad`
    with torch.no_grad():
        w -= w.grad * step
        b -= b.grad * step

    # обнулим градиенты, мы не хотим их накапливать в данном случае
    w.grad = None
    b.grad = None

# посмотрим, уменьшилось ли значение ошибки
print(loss)
```
